```python
data = pd.read_csv("sample2.csv", sep=', ', engine='python')
```

```python
data.head()
```

```python
gb_size_data = data.groupby(['Team', 'Position']).size()
gb_size_data
```

```python
team_players_data = gb_size_data.unstack('Position')
team_players_data
```

```python
team_players_data.fillna(.0, inplace=True)
team_players_data
```

# Advanced indexing

```python
team_players_data_stacked = team_players_data.stack()
team_players_data_stacked
```

```python
team_players_data_stacked_mask = team_players_data_stacked > 10
```

```python
team_players_data_stacked_mask
```

```python
team_players_data_index = team_players_data_stacked[team_players_data_stacked_mask].index
team_players_data_index
```

```python
data = pd.read_csv("sample2.csv", sep=', ', engine='python')
data.set_index(['Team', 'Position'], append=True, inplace=True)
data.reset_index([0], inplace=True)
data
```

```python
filtered_data = data.loc[team_players_data_index]
filtered_data
```

```python
gb_filtered_data = filtered_data.groupby(['Team', 'Position'], group_keys=False)
```

```python
def calc_custom_kpi(df: pd.DataFrame) -> pd.Series:
    return (df['Height(inches)'] * 1 + df['Weight(lbs)'] * 1.5 + df['Age'] * 1) / 3
```

```python
custom_kpi_data = gb_filtered_data.apply(calc_custom_kpi)
```

```python
custom_kpi_data
```

```python
mean_team_kpi_min_10_players = custom_kpi_data.groupby('Team').mean()
mean_team_kpi_min_10_players
```
