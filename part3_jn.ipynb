```python
import pandas as pd
```

```python
data = pd.read_csv("sample2.csv", sep=', ', engine='python')
```

```python
data.head()
```

```python
data['Position'].value_counts()
```

```python
grouped_data = data.groupby('Position')
```

```python
type(grouped_data)
```

```python
grouped_data.get_group('Designated Hitter').head()
```

```python
for group_key, group_table in grouped_data:
    print(group_key, type(group_table))
```

```python
grouped_data.head(2)
```

```python
grouped_data.count()
```

```python
grouped_data.mean()
```

```python
grouped_data[['Height(inches)', 'Weight(lbs)', 'Age']].agg(['min', 'max', 'mean', 'count'])
```

```python
grouped_data[['Height(inches)', 'Weight(lbs)', 'Age']].agg({
    'Height(inches)': ['min', 'max'], 
    'Weight(lbs)': ['mean'], 
    'Age': ['min', 'max', 'mean', 'count']})
```
